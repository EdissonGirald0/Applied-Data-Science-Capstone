# **Análisis de ubicaciones de sitios de lanzamiento con Folium**

Encontrar una ubicación óptima para construir un sitio de lanzamiento sin duda implica muchos factores y esperamos poder descubrir algunos de ellos analizando las ubicaciones de los sitios de lanzamiento existentes.

# Tarea 2: Marcar los lanzamientos exitosos/fallidos de cada sitio en el mapa


A continuación, intentemos mejorar el mapa agregando los resultados de lanzamiento de cada sitio y veamos qué sitios tienen tasas de éxito altas.
Recuerde que el marco de datos spacex_df tiene registros de lanzamiento detallados y la columna `class` indica si este lanzamiento fue exitoso o no.


```python
spacex_df.tail(10)
```

A continuación, crearemos marcadores para todos los registros de lanzamiento.
Si un lanzamiento fue exitoso `(class=1)`, entonces usamos un marcador verde y si un lanzamiento falló, usamos un marcador rojo `(class=0)`


Tenga en cuenta que un lanzamiento solo se produce en uno de los cuatro sitios de lanzamiento, lo que significa que muchos registros de lanzamiento tendrán exactamente la misma coordenada. Los grupos de marcadores pueden ser una buena forma de simplificar un mapa que contiene muchos marcadores que tienen la misma coordenada.

Primero, creemos un objeto `MarkerCluster`


_TODO:_ Crear una nueva columna en el marco de datos `launch_sites` llamada `marker_color` para almacenar los colores de los marcadores según el valor de `class`


```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
```

```python
# Function to assign color to launch outcome
def assign_marker_color(launch_outcome):
    if launch_outcome == 1:
        return 'green'
    else:
        return 'red'
    
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)
spacex_df.tail(10)
```

```python
from folium.plugins import MarkerCluster


# Crear un objeto de MarkerCluster
marker_cluster = MarkerCluster()

# Añadir marker_cluster al mapa
mapa.add_child(marker_cluster)

# Iterar sobre cada lanzamiento en el DataFrame y añadir un marcador personalizado
for index, record in spacex_df.iterrows():
    # Obtener las coordenadas y color del marcador
    coordinates = [record['Lat'], record['Long']]
    color = record['marker_color']
    
    # Crear un marcador con el color asignado según el resultado de lanzamiento
    marker = folium.Marker(
        location=coordinates,
        icon=folium.Icon(color=color, icon='rocket', prefix='fa'),  # Icono de 'rocket' para representar el lanzamiento
        popup=folium.Popup(f"Launch Site: {record['Launch Site']}<br>Result: {'Success' if record['class'] == 1 else 'Failure'}")
    )
    
    # Añadir el marcador al clúster
    marker_cluster.add_child(marker)

# Visualizar el mapa
mapa
```

```python
# Importar librerías necesarias

from folium.plugins import MarkerCluster

# Crear el mapa principal
mapa = folium.Map(location=[28.5623, -80.5774], zoom_start=5)

# Crear el cluster de marcadores
marker_cluster = MarkerCluster()

# Añadir el cluster de marcadores al mapa
mapa.add_child(marker_cluster)

# Asumiendo que 'Class' en spacex_df indica éxito (1) o fallo (0)
for index, record in spacex_df.iterrows():
    # Coordenadas del marcador
    coordinate = [record['Lat'], record['Long']]
    
    # Color según el resultado del lanzamiento
    marker_color = 'green' if record['class'] == 1 else 'red'
    
    # Crear el marcador con el icono personalizado
    marker = folium.Marker(
        location=coordinate,
        icon=folium.Icon(color='red', icon="rocket", icon_color=marker_color),
        popup=folium.Popup(f"Launch Site: {record['Launch Site']}<br>Result: {'Success' if record['class'] == 1 else 'Fail'}", max_width=300)
    )
    
    # Añadir el marcador al cluster
    marker_cluster.add_child(marker)

# Visualizar el mapa
mapa
```

# TAREA 3: Calcular las distancias entre un sitio de lanzamiento y sus proximidades


A continuación, debemos explorar y analizar la proximidad de los sitios de lanzamiento.

Primero, agreguemos una `Posición del mouse` en el mapa para obtener las coordenadas del mouse sobre un punto del mapa. De esta manera, mientras explora el mapa, puede encontrar fácilmente las coordenadas de cualquier punto de interés (como el ferrocarril).


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

mapa.add_child(mouse_position)
mapa
```

Ahora acérquese a un sitio de lanzamiento y explore su proximidad para ver si puede encontrar fácilmente algún ferrocarril, autopista, costa, etc. Mueva el mouse a estos puntos y marque sus coordenadas (mostradas en la parte superior izquierda) para indicar la distancia al sitio de lanzamiento.


Puede calcular la distancia entre dos puntos en el mapa basándose en sus valores de "Lat" y "Long" utilizando el siguiente método:

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
    distance
```
